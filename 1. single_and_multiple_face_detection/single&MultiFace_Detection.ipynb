```python
import cv2
import numpy as np
```

```python
face_classifier=cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

```python
image=cv2.imread('/content/player1.jpeg')
```

```python
from google.colab.patches import cv2_imshow

gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
faces=face_classifier.detectMultiScale(gray,1.3,4)

if len(faces) == 0:
  print('No faces Found!')

for(x,y,w,h) in faces:
  cv2.rectangle(image,(x,y),(x+w,y+h),(155,255,0),2)

cv2_imshow(image)
```

```python
image1=cv2.imread('/content/India-ODI-players.jpeg')
```

```python
gray=cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)
new_faces=face_classifier.detectMultiScale(gray,1.3,5)

if len(new_faces) == 0:
  print('No faces Found!')
for (x,y,w,h) in new_faces:
  cv2.rectangle(image1,(x,y),(x+w,y+h),(0,255,255),2)

cv2_imshow(image1)
```
